```python
from datasets import load_dataset
import numpy as np
import matplotlib.pyplot as plt
```

```python
ds = load_dataset("imagefolder", data_dir="../data/trashnet")
```

```python
ds['train'].features
```

```python
carboard_counts = ds['train']['label'].count(0)
glass_counts = ds['train']['label'].count(1)
plastic_counts = ds['train']['label'].count(2)
paper_counts = ds['train']['label'].count(3)
metal_counts = ds['train']['label'].count(4)
trash_counts = ds['train']['label'].count(5)
plt.bar(['cardboard', 'glass', 'plastic', 'paper', 'metal', 'trash'],
        [carboard_counts, glass_counts, plastic_counts, paper_counts, metal_counts, trash_counts])
plt.xlabel('Material Type')
plt.ylabel('Count')
plt.title('Counts of Different Materials in TrashNet Dataset')
plt.show()
```

```python
random_index = np.random.randint(0, carboard_counts)
carboard_image = ds['train'][random_index]['image']
plt.imshow(carboard_image)
plt.axis('off')
plt.title(f'Random Cardboard Image')
```

```python
random_index = np.random.randint(carboard_counts, carboard_counts + glass_counts)
glass_image = ds['train'][random_index]['image']
plt.imshow(glass_image)
plt.axis('off')
plt.title(f'Random Glass Image')
```

```python
random_index = np.random.randint(carboard_counts + glass_counts, carboard_counts + glass_counts + metal_counts)
metal_image = ds['train'][random_index]['image']
plt.imshow(metal_image)
plt.axis('off')
plt.title(f'Random Metal Image')
```

```python
random_index = np.random.randint(carboard_counts + glass_counts + metal_counts, carboard_counts + glass_counts + metal_counts + paper_counts)
paper_image = ds['train'][random_index]['image']
plt.imshow(paper_image)
plt.axis('off')
plt.title(f'Random Paper Image')
```

```python
random_index = np.random.randint(carboard_counts + glass_counts + metal_counts + paper_counts, carboard_counts + glass_counts + metal_counts + paper_counts + plastic_counts)
plastic_image = ds['train'][random_index]['image']
plt.imshow(plastic_image)
plt.axis('off')
plt.title(f'Random Plastic Image')
```

```python
random_index = np.random.randint(carboard_counts + glass_counts + metal_counts + paper_counts + plastic_counts, carboard_counts + glass_counts + metal_counts + paper_counts + plastic_counts + trash_counts)
trash_image = ds['train'][random_index]['image']
plt.imshow(trash_image)
plt.axis('off')
plt.title(f'Random Trash Image')
```
